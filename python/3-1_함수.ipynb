## 함수의 선언과 호출

- 선언
```python
def func_name(parameter1, parameter2, ...):
    code1
    cod2
    ...
    return value
```
- 호출 `func_name(parameter1, parameter2, ...)`
    - *소괄호가 열림 = 함수가 호출됨*

````python
def rectangle(height, width):
    area = height * width
    perimeter = (height + width) * 2

    print(f'사각형의 면적은 {area}이고 둘레는 {perimeter}입니다.')
````

````python
rectangle(50, 30)
rectangle(20, 100)
````
