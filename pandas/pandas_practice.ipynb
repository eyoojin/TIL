- 남성 흡연자 조건 필터링
- tip을 가장 많이 낸 순서로 내림차순 정렬
- 상위 10개 출력

```python
cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'
tips_ms = tips.loc[cond1 & cond2]
tips_ms.sort_values(['tip'], ascending=False).head(10)
```

```python
tips.loc[cond1 & cond2].sort_values(['tip'], ascending=False).head(10)
```

- 식사 인원이 5명 이상인 저녁 테이블
- 흡연자가 없는 주말 테이블
- total_bill, tip 출력

```python
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'
cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat', 'Sun'])

my_tips = tips.loc[cond1 & cond2 & cond3 & cond4]
my_tips[['total_bill', 'tip']]
```

```python
tips.loc[cond1 & cond2 & cond3 & cond4][['total_bill', 'tip']]
tips.loc[cond1 & cond2 & cond3 & cond4, ['total_bill', 'tip']]
```

# 05. missing value

- 남자 승객의 결측치는 남자 승객의 평균
- 여자 승객의 결측치는 여자 승객의 평균
- 결측치를 모두 채우고 나서 모든 사람의 평균

```python
df = sns.load_dataset('titanic')
df_copy = df.copy()
```

```python
cond1 = df_copy['sex'] == 'male'
cond2 = df_copy['sex'] == 'female'

df_copy.loc[cond1, 'age'] = df_copy.loc[cond1, 'age'].fillna(df_copy.loc[cond1, 'age'].mean())
df_copy.loc[cond2, 'age'] = df_copy.loc[cond2, 'age'].fillna(df_copy.loc[cond2, 'age'].mean())

print(df_copy.loc[cond1, 'age'].mean())
print(df_copy.loc[cond2, 'age'].mean())

df_copy['age'].mean()
```
