```python
e_m = df['embarked'].mode()
e_m[0]
```

```python
df['embarked'] = df['embarked'].fillna(e_m[0])
```

```python
assert 0 == df['embarked'].isnull().sum()
```

```python
m_m = round(df.loc[df['sex'] == 'male', 'age'].mean(), 4)
f_m = round(df.loc[df['sex'] == 'female', 'age'].mean(), 4)
```

```python
print(m_m)
print(f_m)
```

```python
df.loc[df['sex'] == 'male', 'age'] = df.loc[df['sex'] == 'male', 'age'].fillna(m_m)
df.loc[df['sex'] == 'female', 'age'] = df.loc[df['sex'] == 'female', 'age'].fillna(f_m)
```

```python
assert df.groupby('sex')['age'].mean()['female'].round(4) == 27.9157
assert df.groupby('sex')['age'].mean()['male'].round(4) == 30.7266
assert df['age'].isnull().sum() == 0
```
